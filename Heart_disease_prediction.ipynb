```python
import pandas as pd

df = pd.read_csv("dataset_heart.csv")  # replace with your file name
```

```python
df.head()
```

```python
from pycaret.classification import *

# This initializes the environment and does all preprocessing automatically
clf_setup = setup(data=df,
                  target='heart disease',
                  session_id=123,
                  train_size=0.8)
```

```python
best_model = compare_models()
```

```python
evaluate_model(best_model)
```

```python
predictions = predict_model(best_model)
```

```python
save_model(best_model, 'Heart_disease_prediction_model')
```
